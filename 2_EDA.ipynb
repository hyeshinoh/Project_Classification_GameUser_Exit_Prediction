### (5) Party에 많이 참여할수록 잔존 가능성이 높다
- party_count threshold에 따라 각 label의 비중이 달라짐
- 특히 약 75분위수인 170회 이상으로 넘어가면 retained의 비중이 높아짐

```python
counts_base = train_merge[train_merge["party_cnt"] >= 0].groupby('label').size().reset_index(name='id_counts')
counts_100 = train_merge[train_merge["party_cnt"] >= 100].groupby('label').size().reset_index(name='id_counts')
counts_200 = train_merge[train_merge["party_cnt"] >= 200].groupby('label').size().reset_index(name='id_counts')
counts_500 = train_merge[train_merge["party_cnt"] >= 500].groupby('label').size().reset_index(name='id_counts')
```

```python
fig = plt.figure(figsize=(20, 5))
fig.suptitle("party count threshold 변화에 따른 각 class의 비중", fontsize=20)
ax1 = plt.subplot(1, 4, 1)
ax1.set_title("전체 data의 각 class 비중")
sns.barplot(x = 'label', y = 'id_counts', data = counts_base, ax = ax1)

ax2 = plt.subplot(1, 4, 2)
ax2.set_title("party_cnt > 100 인 경우 각 class의 비중")
sns.barplot(x = 'label', y = 'id_counts', data = counts_100, ax = ax2)

ax3 = plt.subplot(1, 4, 3)
ax3.set_title("party_cnt > 200 인 경우 각 class의 비중")
sns.barplot(x = 'label', y = 'id_counts', data = counts_200, ax = ax3)

ax4 = plt.subplot(1, 4, 4)
ax4.set_title("party_cnt > 500 인 경우 각 class의 비중")
sns.barplot(x = 'label', y = 'id_counts', data = counts_500, ax = ax4)
```
